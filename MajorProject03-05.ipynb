### Complex hist plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generate random data for 3 datasets
data1=np.random.normal(loc=0,scale=1,size=1000)
data2=np.random.normal(loc=2,scale=0.5,size=1000)
data3=np.random.normal(loc=-2,scale=1.5,size=1000)

#figure size
plt.figure(figsize=(10,6))

#plot histogram
plt.hist(data1,bins=30,alpha=0.5,label='Data1(mean=0,std=1)',color='m',edgecolor='k',density=True)
plt.hist(data2,bins=30,alpha=0.5,label='Data2(mean=2,std=0.5)',color='b',edgecolor='k',density=True)
plt.hist(data3,bins=30,alpha=0.5,label='Data3(mean=-2,std=0.5)',color='r',edgecolor='k',density=True)

#add labels
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Histogram with multiple datasets')

#add grid
plt.grid(True,linestyle=':',linewidth=0.7)

#add legend
plt.legend()

#show plot
plt.tight_layout()
plt.show()
```

### Pie Chart

```python
#data 
labels=['Apple','Banana','Cherry','Date']
sizes=[30,25,25,20]
colors=['red','yellow','pink','brown']
explode=(0.1,0,0,0)

#create the pie chart
plt.figure(figsize=(5,5))
plt.pie(sizes,labels=labels,colors=colors,explode=explode,autopct='%1.1f%%',shadow=True,startangle=140)

#equal aspect ratio ensures the pie is drawn as a circle
plt.axis('equal')
plt.title('Fruit Distribution')
plt.show()
```

### Bar chart

```python
#data
fruits=['Apple','Banana','Mango','Grapes']
quantities=[40,25,30,15]

#create bar chart
plt.bar(fruits,quantities,color='red',edgecolor='k')
plt.grid(True,linestyle=':',linewidth=0.5)
#labels
plt.xlabel('Fruits')
plt.ylabel('Quantities')
plt.title('Fruit Stock')
plt.show()
```

### complex bar chart

```python
#data
languages=['Python','Java','C++','Ruby','JavaScript']
popularity=[85,75,60,40,70]
colors=('r','m','w','pink','yellow')

#create bar chart with colors
plt.figure(figsize=(8,5))
bars=plt.bar(languages,popularity,color=colors,edgecolor='k')

#labels and title
plt.xlabel('Programming language')
plt.ylabel('Popularity(%)')
plt.title('Programming language popularity')
plt.ylim(0,100)

#show grid and plot
plt.grid(axis='y',linestyle='--',alpha=0.6)
plt.tight_layout()
plt.show()
```

### Box plot

```python
#sample data
np.random.seed(42)
data=[np.random.normal(loc=70,scale=10,size=100),
      np.random.normal(loc=65,scale=15,size=100),
      np.random.normal(loc=80,scale=5,size=100),
      np.random.normal(loc=75,scale=12,size=100)]

#labels for each class
labels=['Class A','Class B','Class C','Class D']

#create the box plot
plt.boxplot(data,labels=labels,patch_artist=True,boxprops=dict(facecolor='lightblue',color='blue'),medianprops=dict(color='r'),whiskerprops=dict(color='green'),capprops=dict(color='yellow'))

#add title and labels
plt.ylabel('Scores')
plt.title('Test Scores by Class')
plt.grid(True)
plt.show()
```

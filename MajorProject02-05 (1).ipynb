### Visualization

#### Line plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,100,25)
y=np.linspace(0,1000,25)
```

```python
plt.plot(x,y)
plt.show()
```

```python
#multiple plots
x=np.linspace(1,10,100)
plt.figure(figsize=(8,8))

#subplot1
plt.subplot(2,2,1)
plt.title("Linear")
plt.plot(x,x)

#subplot2
plt.subplot(2,2,2)
plt.title("Cubic")
plt.plot(x,x**3)

#subplot3
plt.subplot(2,2,3)
plt.title("Log")
plt.plot(x,np.log(x))

#subplot4
plt.subplot(2,2,4)
plt.title("Square")
plt.plot(x,x**2)

plt.show()
```

```python
#both in one plot
x=np.linspace(0,100,25)
y=np.linspace(0,1000,25)
plt.plot(x,x,label='x vs x',color='r')
plt.plot(x,y,label='x vs x**3',color='g')
plt.title('x vs x,x vs x**3')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#complex line plot
x=np.linspace(0,10,100)
#multiple functions
y1=np.sin(x)
y2=np.cos(x)
y3=np.sin(x)+np.cos(x)

#create the plot
plt.figure(figsize=(10,6))

#plot lines with different styles
plt.plot(x,y1,color='r',linestyle='--',marker='o',linewidth='2',markersize=5,label='sin(x)')
plt.plot(x,y2,color='m',linestyle='-.',marker='s',linewidth='2',markersize=5,label='cos(x)')
plt.plot(x,y3,color='k',linestyle='-',linewidth='2',label='sin(x)+cos(x)')

#fill between two lines
plt.fill_between(x,y1,y2,color='gray',alpha=0.2,label='Area between sin(x) and cos(x)')

#add grid
plt.grid(True,linestyle=':',linewidth=0.2)

#add title and axis labels
plt.title('complex line plot')
plt.xlabel('x axis')
plt.ylabel('y axis')

#custom ticks
plt.xticks(np.arange(0,11,1))
plt.yticks(np.arange(-2,2.5,0.5))

#annotation
plt.annotate('Intersection',xy=(2.2,0),xytext=(3,0.5),arrowprops=dict(facecolor='black',shrink=0.05))

#add legend
plt.legend(loc='upper right')

#show plot
plt.tight_layout()
plt.show()
```

### Scatter plot

```python
#sample data
x=np.linspace(0,10,25)
y=x**2
#scatter plot
plt.scatter(x,y,s=100,marker='*',edgecolor='r',color='g',label='y=**x')
#add legend
plt.legend()
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Scatter plot')
plt.show()
```

```python
#data
x=np.random.rand(100)
y=np.random.rand(100)
#scatter plot
plt.scatter(x,y,label='scatter',s=10,marker='o',edgecolor='k',color='m')
plt.title('Scatter plot')
plt.show()
```

```python
#complex scatter plot
np.random.seed(42)
x=np.random.rand(100)*10
y=np.random.rand(100)*10
colors=np.sqrt(x**2,y**2)
sizes=np.random.rand(100)*300

#scatter plot
scatter=plt.scatter(x,y,c=colors,
                    s=sizes,
                    cmap='viridis',
                    alpha=0.7,
                    edgecolor='black',
                    marker='o',
                    label='Data Points')
#add colors
cbar=plt.colorbar(scatter)
cbar.set_label('Distance from origin')

#plot labels
plt.xlabel('feature x')
plt.ylabel('feature y')
plt.title('Complex scatter plot')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

### Histogram

```python
##generate
data=np.random.rand(100)

#create a histogram
plt.hist(data,bins=10,color='skyblue',edgecolor='k',alpha=1.0)

#add labels and title
plt.title('Histogram of random data')
plt.xlabel('Value')
plt.ylabel('Frequency')

#show the plot
plt.grid(True)
plt.show()
```

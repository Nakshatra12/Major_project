```python
age_salary = df.groupby('age')['Anual_Salary'].mean().sort_index()
plt.figure(figsize=(16,10))
plt.plot(age_salary.index,age_salary.values,color='green')
plt.xlabel('Age')
plt.ylabel('Average Salary')
plt.title('Average Salary by Age')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#line plot'
age_children = df.groupby('age')['children'].mean().sort_index().round()
plt.figure(figsize=(16,10))
plt.plot(age_children.index,age_children.values,color='green')
plt.xlabel('Age')
plt.ylabel('Children')
plt.title('Children by Age')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#line plot
age_Hospital_expenditure = df.groupby('age')['Hospital_expenditure'].mean().sort_index()
plt.figure(figsize=(16,10))
plt.plot(age_Hospital_expenditure.index,age_Hospital_expenditure.values,color='green')
plt.xlabel('Age')
plt.ylabel('Hospital_expenditure')
plt.title('Hospital_expenditure by Age')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#scatter plot
plt.figure(figsize=(10,8))
plt.scatter(df['age'],df['bmi'],s=40,color='r',edgecolor='k',marker='*',label='scatter')
plt.xlabel('Age')
plt.ylabel('Bmi')
plt.title('Age Vs Bmi')
plt.legend()
plt.show()
```

```python
#scatter plot
plt.figure(figsize=(15,9))
plt.scatter(df['age'],df['Claim_Amount'],s=100,marker='*',color='skyblue',edgecolor='k',label='Scatter')
plt.xlabel('Age')
plt.ylabel('Claim_Amount')
plt.title('Age Vs Claim Amount')
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
#histogram
plt.figure(figsize=(10,5))
plt.hist(df['age'],color='gray',edgecolor='black',label='hist',alpha=0.6)
plt.xlabel('Age')
plt.legend()
plt.grid(linestyle=':',linewidth=1.5)
plt.show()
```

```python
#hist
plt.figure(figsize=(2,1))
plt.hist(df['children'])
plt.show()
```

```python
#pie chart
region_counts=df['region'].value_counts()
#optional explode
explode=[0.1]*len(region_counts)
plt.figure(figsize=(6,6))
plt.pie(region_counts,labels=region_counts.index,explode=explode,autopct='%1.1f%%',shadow=True,startangle=360)
plt.title('Region distribution')
plt.axis('equal')
plt.show()
```

```python
#pie chart
smoker_counts=df['smoker'].value_counts()
#optional explode
explode=[0.1]*len(smoker_counts)
plt.figure(figsize=(6,6))
plt.pie(smoker_counts,labels=smoker_counts.index,explode=explode,autopct='%1.1f%%',shadow=True,startangle=360)
plt.title('smoker distribution')
plt.axis('equal')
plt.show()
```

```python
#bar
plt.figure(figsize=(10,6))
region=df['region'].value_counts()
plt.bar(region.index,region)
plt.xlabel('region')
plt.ylabel('count')
plt.title('region')
plt.show()
```

```python
#bar
plt.figure(figsize=(10,6))
sex=df['sex'].value_counts()
plt.bar(sex.index,sex)
plt.xlabel('sex')
plt.ylabel('count')
plt.title('gender')
plt.show()
```

### netflix dataset

```python
#another dataset

df1=pd.read_csv("C:\\Users\\srini\\OneDrive\\Desktop\\NFLX.csv")
```

```python
df1
```

### lineplot

```python
df1['Date']=pd.to_datetime(df1['Date'])
#line plot
plt.figure(figsize=(20,8))
plt.plot(df1['Date'],df1['Adj Close'],color='r')
plt.show()
```

```python

#get 10 evenly spaced dates from the dataset
tick_dates=df1['Date'].iloc[::len(df1)//10][:10]
plt.figure(figsize=(20,8))
plt.plot(df1['Date'],df1['Adj Close'],color='r')
plt.xlabel('Date')
plt.ylabel('Adjusted Close Price')
plt.title('Adjusted Close Price over time')

#use only 10 ticks
plt.xticks(tick_dates,rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Scatter plot

```python
#scatter plot
plt.figure(figsize=(10,6))
plt.scatter(df1['Open'],df1['Close'],color='red',edgecolor='black',marker='o',s=50,label='Scatter')
plt.xlabel('Open')
plt.ylabel('Close')
plt.title('Open & Close')
plt.legend()
plt.grid(True)
plt.show()
```

### Histogram

```python
#histogram
plt.figure(figsize=(10,8))
plt.hist(df1['Open'],bins=30,alpha=0.5,color='gray',edgecolor='k')
plt.xlabel('Open')
plt.ylabel('Frequency')
plt.show()
```

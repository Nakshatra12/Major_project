### vizualization using seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading dataset
tips=sns.load_dataset('tips')
tips
```

```python
#loading dataset
tips=sns.load_dataset('tips')

#create hist plot
sns.histplot(data=tips,x='total_bill',bins=20,kde=True,color='red',edgecolor='black')

#show the plot with labels and title
plt.title('Distribution of Total Bills')
plt.xlabel('Total Bills ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
### create histogram with hue
sns.histplot(data=tips,x='tip',kde=True,hue='smoker',bins=30)

#show the plot with labels and title
plt.title('Distribution of Total Bills')
plt.xlabel('Total tips')
plt.ylabel('frequency')
plt.show()
```

```python
### create histogram with hue with custom palatte
sns.histplot(data=tips,x='total_bill',kde=True,hue='smoker',bins=30,palette={'Yes':'red','No':'green'})

#show the plot with labels and title
plt.title('Distribution of Total Bills')
plt.xlabel('Total bills')
plt.ylabel('frequency')
plt.show()
```

```python
### create histogram with hue with custom palatte
sns.histplot(data=tips,x='total_bill',kde=True,hue='smoker',bins=30,palette={'Yes':'red','No':'green'})

#show the plot with labels and title
plt.title('Distribution of Total Bills')
plt.xlabel('Total bills ($)')
plt.ylabel('frequency')
plt.show()
```

```python
#hist with numerical & catg col
#set a style
sns.set(style='whitegrid')
g=sns.FacetGrid(tips,col='time',height=4,aspect=1.2,hue='smoker')
g.map(sns.histplot,'total_bill',bins=20,kde=True,edgecolor='black',alpha=0.5)
g.add_legend()
g.set_titles(col_template="{col_name}Time")
g.set_axis_labels('Total_bill','count')
g.fig.suptitle('Distribution of total bills by time and smokers')
g.fig.subplots_adjust(top=0.85)
plt.show()
```

```python
#define custom color palette

custom_palette={'Yes':'red','No':'green'}

#create facetgrid without hue directly in facetgrid
g=sns.FacetGrid(tips,col='time',height=4,aspect=1.2)
g.map_dataframe(sns.histplot,
                x='total_bill',
                hue='smoker',
                bins=30,
                kde=True,
                palette=custom_palette,
                edgecolor='k',
                label='time',
                alpha=0.7)
#set titles and labels
g.set_titles(col_template="{col_name}Time")
g.set_axis_labels("Total bill","count")
plt.subplots_adjust(top=0.85)
g.fig.suptitle('Distribution of total bill (custom colors)')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches

# Adjust subplot layout and add title
plt.subplots_adjust(top=0.88)
fig = plt.gcf()
fig.suptitle("Distribution of Total Bill (Custom Colors)")

# Manually add legend with custom handles
handles = [
    mpatches.Patch(color="green", label="YES"),
    mpatches.Patch(color="red", label="NO")
]
plt.legend(title="Smoker", handles=handles, loc="upper right")
plt.show()
```

```python
# Box plot
sns.barplot(data=tips, x="day", y="total_bill")
plt.title("Average Total Bill")
plt.xlabel("Day of the Week")
plt.ylabel("Average Total Bill ($)")
plt.show()
```

```python
# Bar plot with hue
sns.barplot(data=tips, x="day", y="total_bill", hue="smoker", palette="pastel")
plt.title("Average Total Bill by Day & Smoker Status")
plt.xlabel("Day")
plt.ylabel("Average Total Bill ($)")
plt.show()
```

```python
# Bar plot with hue
sns.barplot(data=tips, x="day", y="total_bill", hue="smoker", palette="pastel")
plt.title("Average Total Bill by Day & Smoker Status")
plt.xlabel("Day")
plt.ylabel("Average Total Bill ($)")
plt.show()
```

```python
# Count plot (basic)
sns.countplot(data=tips, x="day")
plt.title("Count of Entries by Day")
plt.xlabel("Day")
plt.ylabel("Count")
plt.show()
```

```python
# Count plot with order and palette
sns.countplot(data=tips, x="day", order=["Sun", "Sat", "Fri", "Thur"], palette="pastel")
plt.title("Count of Entries by Day (Ordered)")
plt.xlabel("Day")
plt.ylabel("Count")
plt.show()
```

```python
# Bar plot with custom order and palette
sns.barplot(
    data=tips, x="day", y="total_bill", hue="smoker",
    order=["Sun", "Sat", "Fri", "Thur"],
    palette={"Yes": "r", "No": "b"}
)
plt.title("Average Total Bill by Day & Smoker Status (Custom Palette)")
plt.xlabel("Day")
plt.ylabel("Average Total Bill ($)")
plt.show()
```

```python
# Basic scatter plot
sns.scatterplot(data=tips, x="total_bill", y="tip",hue='smoker')
plt.title("Scatter Plot of Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
# line plot  
# load the flights dataset  
flights = sns.load_dataset("flights")

# preview the data  
flights.head()
```

```python
# line plot  
sns.lineplot(data=flights, x="month", y="passengers",label='year')  
plt.title("Monthly Airline Passengers")  
plt.xlabel('month')
plt.ylabel('passengers')  
plt.legend(title="year", bbox_to_anchor=(1.05, 1), loc="upper left")  
plt.tight_layout()
plt.show()
```

```python
# palette & hue  
sns.lineplot(data=flights, x="month", y="passengers", hue="year", palette="tab10")
plt.title("Monthly Airline Passengers")  
plt.xlabel('month')
plt.ylabel('passengers')  
plt.legend(title="year", bbox_to_anchor=(1.05, 1), loc="upper left")  
plt.tight_layout()
plt.show()
```

```python
# set  
import pandas as pd  
month_order = ["Jan", "Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
flights['month']=pd.Categorical(flights['month'],categories=month_order,ordered=True)
#now re-plot
sns.lineplot(data=flights,x='month',y='passengers',hue='year',palette='Set3')
plt.title('Monthly airline passengers(ordered months)')
plt.xlabel('month')
plt.ylabel('Passengers')
plt.legend(title="year", bbox_to_anchor=(1.05, 1), loc="upper left")  
plt.tight_layout()
plt.show()
```
